```python
fig = plt.figure(figsize = (4,4))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y) 
ax.plot(y,x)
ax.set_title('1st Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
 
```

```python
fig = plt.figure(figsize = (4,4))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X Sqred') 
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.plot(x,x**3,label = 'X Cubed') 

ax.legend(loc = 2)
```

```python
### Location String	          Location Code
##### 'best'                      0
##### 'upper right'	              1
##### 'upper left'	              2
##### 'lower left'	              3
##### 'lower right'	              4
##### 'right'	                  5
##### 'center left'	              6
##### 'center right'	          7
##### 'lower center'	          8
##### 'upper center'	          9
##### 'center'	                  10
```

```python
fig = plt.figure(figsize = (4,4))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label = 'X Sqred',color = 'purple', linewidth = 5) 

ax.plot(x,x**3,label = 'X Cubed',lw = 3, alpha = 0.5) 

ax.legend(loc = 'upper right')
```

```python
fig = plt.figure(figsize = (4,4))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label = 'X Sqred',color = 'purple', linestyle = '--') 

ax.plot(x,x**3,label = 'X Cubed',lw = 3, alpha = 0.5,ls = '-.') 

ax.legend(loc = 'upper right')
```

# Markers

```python
fig = plt.figure(figsize = (4,4))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**3,label = 'X Cubed',marker = 'o',lw = 2, alpha = 0.6,ls = '--') 

ax.legend(loc = 'upper left')
```

```python
fig = plt.figure(figsize = (4,4))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**3,label = 'X Cubed',marker = 'o',markersize = 10,lw = 2, alpha = 0.5,ls = '--') 
```

```python
fig = plt.figure(figsize = (4,4))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**3,label = 'X Cubed',marker = 'o',markersize = 20,markerfacecolor='green',
        markeredgewidth=2,markeredgecolor='yellow',lw = 2, ls = '--') 
```

```python
fig = plt.figure(figsize = (4,4))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**3,label = 'X Cubed',lw = 2, ls = '--') 

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```

### other examples

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.set_title('title')
ax.set_xlabel('x')
ax.set_ylabel('y')

ax.plot(x,y)
```

```python
fig1 = plt.figure()

ax1 = fig1.add_axes([0,0,1,1])
ax2 = fig1.add_axes([0.2,0.5,.2,.2])

ax1.plot(x,y)
ax2.plot(y,x)
```

```python
fig1 = plt.figure()
ax0 = fig1.add_axes([0,0,1,1])
ax1 = fig1.add_axes([0.15,0.4,.4,.4])

ax0.set_xlabel('X')
ax0.set_ylabel('Z')

ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax1.set_ylim(30,50)
ax1.set_xlim(20,22)

ax0.plot(x,z)
ax1.plot(x,y)
```
